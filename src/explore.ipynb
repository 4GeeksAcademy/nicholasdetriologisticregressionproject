```python
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome', ]

fig, axis = plt.subplots(2, 5, figsize = (20, 10))

# Creating a multiple figure with histograms and box plots
sns.histplot(ax = axis[0, 0], data = total_data, x = "job")
sns.histplot(ax = axis[0, 1], data = total_data, x = "marital")
sns.histplot(ax = axis[0, 2], data = total_data, x = "education")
sns.histplot(ax = axis[0, 3], data = total_data, x = "default")
sns.histplot(ax = axis[0, 4], data = total_data, x = "housing")
sns.histplot(ax = axis[1, 0], data = total_data, x = "loan")
sns.histplot(ax = axis[1, 1], data = total_data, x = "contact")
sns.histplot(ax = axis[1, 2], data = total_data, x = "month")
sns.histplot(ax = axis[1, 3], data = total_data, x = "day_of_week")
sns.histplot(ax = axis[1, 4], data = total_data, x = "poutcome")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
numerical_features = ['Age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

fig, axis = plt.subplots(4, 2, figsize = (10, 10), gridspec_kw={'height_ratios': [6, 1, 6, 1]})

# Creating a multiple figure with histograms and box plots
sns.histplot(ax = axis[0, 0], data = total_data, x = "age").set(xlabel = None)
sns.boxplot(ax = axis[1, 0], data = total_data, x = "age")
sns.histplot(ax = axis[0, 1], data = total_data, x = "duration").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[1, 1], data = total_data, x = "duration")

sns.histplot(ax = axis[2, 0], data = total_data, x = "campaign").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[3, 0], data = total_data, x = "campaign")
sns.histplot(ax = axis[2, 1], data = total_data, x = "pdays").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[3, 1], data = total_data, x = "pdays")


# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
